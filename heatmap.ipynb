```python
import numpy as np
import matplotlib.pyplot as plt

harvest = np.array([[1, 2, 3],
                    [3, 4, 5],
                    [4, 5, 6]])
labels = ["aaaaaaaa", "bbbbbbb", "cccccccccc"]
fig, ax = plt.subplots()
ax.set_xticks(np.arange(len(harvest)), labels=labels, rotation=45, rotation_mode="anchor", ha="right")
ax.set_yticks(np.arange(len(labels)), labels=labels)
for i in range(len(harvest)):
    for j in range(len(harvest[i])):
        ax.text(j, i, harvest[i, j], ha="center", va = "center", color="w")
a = ax.imshow(harvest, cmap="plasma")
fig.colorbar(a)
fig.show()
```
